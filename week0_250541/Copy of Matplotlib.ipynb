# #**Plots**

Plot the following plots using matplotlib, you many generate your own data or get it from the internet:

1. Histogram
2. Scatter Plot
3. Boxplot
4. Pie charts

(Bonus - Read up on the use-case of histograms in EDA and ML in general, they are VERY useful! Also, look up **Violin plots** if you have extra time)


```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
fig = plt.figure()
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
np.random.seed(42)
data_normal = np.random.normal(100, 15, 1000)  # Normal distribution
data_skewed = np.random.exponential(2, 1000)   # Skewed distribution
categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 45, 12, 67, 33]
scatter_x = np.random.randn(100)
scatter_y = 2 * scatter_x + np.random.randn(100) * 0.5
# Histogram (top-left)
axs[0, 0].hist(data_normal, bins=30, alpha=0.7, color='skyblue', edgecolor='black')
axs[0, 0].set_title('Histogram')

# Scatter Plot (top-right)
axs[0, 1].scatter(scatter_x, scatter_y, alpha=0.6, color='coral', s=50)
axs[0, 1].set_title('Scatter Plot')
axs[0, 1].set_xlabel('X')
axs[0, 1].set_ylabel('Y')

# Boxplot (bottom-left)
axs[1, 0].boxplot([data_normal, data_skewed], labels=['Normal', 'Skewed'])
axs[1, 0].set_title('Boxplot')
axs[1, 0].set_ylabel('Value')

# Pie Chart (bottom-right)
axs[1, 1].pie(values, labels=categories, autopct='%1.1f%%', startangle=90)
axs[1, 1].set_title('Pie Chart')
```
