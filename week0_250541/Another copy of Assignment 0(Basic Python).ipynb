(i) Check if matrix multiplication is valid

```python
#Q3(i)
# if (no. of columns of matrix_1) = (no. of rows of matrix_2)
#then matrix multiplication is valid
if matrix_1.shape[1] == matrix_2.shape[0]:
  print("matrix multiplication is valid")
else:
  print("matrix multiplication is invalid")

```

(ii) Using loops (without using numpy)

```python
#Q3(ii)
rows_1 = len(matrix_1)
coln_1 = len(matrix_1[0])
rows_2 = len(matrix_2)
coln_2 = len(matrix_2[0])

if coln_1 != rows_2:
  print ("matrix multiplication is not valid")
else:
  result = [[0 for _ in range(coln_2)] for _ in range (rows_1)]



for i in range(rows_1):
  for j in range(coln_2):
    for k in range(coln_1):
      result[i][j] += matrix_1[i][k] * matrix_2[k][j]

print("result of matrix multiplication: ")
for row in result:
  print(row)




```
