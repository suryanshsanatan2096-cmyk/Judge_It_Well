**Importing pipeline from transformers**

Using pipeline we can use several pre-trained models

```python
from transformers import pipeline
```

**Model for Sentiment Analysis**

```python
classifier = pipeline('sentiment-analysis')
result = classifier("I Love BCS project Judge-it-well")
print(result)
```

**Model for Text generation**

```python
generator = pipeline('text-generation')
result = generator("Indian institue of Technology Kanpur is one of ")
print(result)
```

**Model for translation**

```python
translator = pipeline('translation_en_to_fr')
result = translator("BCS stands for Brain and cognitive science")
print(result)
```
