**Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
fig1=plt.figure()
ax1=fig1.add_axes([0,0,1,1])
ax2=fig1.add_axes([0.2,0.5,.4,.4])
z=x**2
ax1.plot(x,z)
ax2.set_title("zoom")
ax2.set_xlabel("X")
ax2.set_ylabel("Y")
ax1.set_xlabel("X")
ax1.set_ylabel("Z")
ax2.plot(x,y)
y=2*x
ax2.set_xlim([20,22])
ax2.set_ylim([30,50])
plt.show()
```

## Exercise 4

**Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
figure,axis=plt.subplots(nrows=1,ncols=2)
```

**Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
figure,axis=plt.subplots(nrows=1,ncols=2)
axis[0].plot(x,y,color="Blue",linestyle="--")
axis[1].plot(x,z,color="Red")
```

**See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.**

```python
figure,axis=plt.subplots(nrows=1,ncols=2,figsize=(12,2))
axis[0].plot(x,y,color="Blue",linestyle="--")
axis[1].plot(x,z,color="Red")
```

Plot the following plots using matplotlib, you many generate your own data or get it from the internet:

(Bonus - Read up on the use-case of histograms in EDA and ML in general, they are VERY useful! Also, look up **Violin plots** if you have extra time)


```python
import numpy as np
import matplotlib.pyplot as plt
##draw a histogram
data=np.random.randn(1000)
plt.hist(data,color="pink")
X=np.random.randint(1,100,size=(100))
Y=np.random.randint(1,100,size=(100))
plt.figure(figsize=(3,3))
plt.scatter(X,Y,color="red")##scattered plot
plt.figure(figsize=(3,3))
plt.boxplot(data)##boxplot
plt.figure()
data1=np.random.randint(1,100,size=(100))
labels=["A","B","C","D"]
plt.pie(data1,labels=labels,autopct="%1.1f%%")##piechart
plt.show()
```
