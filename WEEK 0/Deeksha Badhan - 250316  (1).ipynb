```python
##mean of x
sum=0
X=[1,2,3,4,5]
b=len(X)
for n in X:
  sum=sum+n
meanofx=sum/b
print(meanofx)
```

```python
##mean of y
sum1=0
Y=[2,4,5,4,5]
a=len(Y)
for n in Y:
  sum1=sum1+n
meanofy=sum1/a
print(meanofy)
```

```python
##slope of the data
X=[1,2,3,4,5]
Y=[2,4,5,4,5]
n=0
d=0
for i in range(len(X)):
  n=n+((X[i]-meanofx)*(Y[i]-meanofy))
for j in range(len(X)):
  d=d+((X[j]-meanofx)**2)
slope=n/d
print("The slope is:",slope)
```

```python
##y intercept
c=meanofy-(slope*meanofx)
print("The y intercept is:",c)
```
