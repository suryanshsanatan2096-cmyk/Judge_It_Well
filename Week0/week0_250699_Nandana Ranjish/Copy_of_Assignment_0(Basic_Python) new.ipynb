Compute:

intercept (c)

predicted values

No libraries allowed.(sklearn or numpy)


```python
X=[1,2,3,4,5]
sum=0
for i in X:
  sum+=i
meanx=sum/len(X)
print(meanx)
```

```python
Y=[2,4,5,4,5]
sum=0
for i in Y:
  sum+=i
meany=sum/len(Y)
print(meany)
```

```python
num=0
den=0
for i in range(len(X)):
  num+=(X[i]-meanx)*(Y[i]-meany)
  den+=(X[i]-meanx)*(X[i]-meanx)
m=num/den
print(m)
```
