**Q.3**

(i) Check if matrix multiplication is valid

```python
#your code here# Q3(i): Check if matrix multiplication is valid
# Matrix multiplication A @ B is valid if columns of A == rows of B

def is_matrix_multiplication_valid(matrix_A, matrix_B):
    rows_A, cols_A = len(matrix_A), len(matrix_A[0])
    rows_B, cols_B = len(matrix_B), len(matrix_B[0])

    is_valid = (cols_A == rows_B)

    print(f"Matrix A shape: {rows_A} x {cols_A}")
    print(f"Matrix B shape: {rows_B} x {cols_B}")
    print(f"Valid for multiplication: {is_valid}")
    if is_valid:
        print(f"Result shape will be: {rows_A} x {cols_B}")

    return is_valid

# Test cases
A1 = [[1, 2, 3], [4, 5, 6]]  # 2x3
B1 = [[1, 2], [3, 4], [5, 6]]  # 3x2
print("Test 1:")
is_matrix_multiplication_valid(A1, B1)
print()

A2 = [[1, 2], [3, 4]]  # 2x2
B2 = [[5, 6, 7], [8, 9, 10]]  # 2x3
print("Test 2:")
is_matrix_multiplication_valid(A2, B2)
print()

A3 = [[1, 2, 3]]  # 1x3
B3 = [[1], [2]]  # 2x1 - invalid
print("Test 3:")
is_matrix_multiplication_valid(A3, B3)
```

(ii) Using loops (without using numpy)

```python
#your code here# Q3(ii): Matrix multiplication using loops (without numpy)

def matrix_multiply_loops(A, B):
    """
    Multiply two matrices A and B using nested loops
    A: m x n, B: n x p => Result: m x p
    """
    # Get dimensions
    rows_A = len(A)
    cols_A = len(A[0])
    rows_B = len(B)
    cols_B = len(B[0])

    # Check if multiplication is valid
    if cols_A != rows_B:
        print("Error: Matrix multiplication not possible")
        return None

    # Initialize result matrix with zeros
    result = [[0 for _ in range(cols_B)] for _ in range(rows_A)]

    # Perform multiplication: result[i][j] = sum of A[i][k] * B[k][j]
    for i in range(rows_A):
        for j in range(cols_B):
            for k in range(cols_A):
                result[i][j] += A[i][k] * B[k][j]

    return result

def print_matrix(matrix, name):
    """Print matrix in readable format"""
    print(f"{name}:")
    for row in matrix:
        print([f"{val:6.1f}" for val in row])
    print()

# Test case 1: 2x3 @ 3x2 = 2x2
A1 = [[1, 2, 3], [4, 5, 6]]
B1 = [[7, 8], [9, 10], [11, 12]]
result1 = matrix_multiply_loops(A1, B1)
print("Test 1: 2x3 matrix @ 3x2 matrix")
print_matrix(A1, "Matrix A")
print_matrix(B1, "Matrix B")
if result1:
    print_matrix(result1, "Result (A @ B)")

# Test case 2: 3x2 @ 2x3 = 3x3
A2 = [[1, 2], [3, 4], [5, 6]]
B2 = [[7, 8, 9], [10, 11, 12]]
result2 = matrix_multiply_loops(A2, B2)
print("Test 2: 3x2 matrix @ 2x3 matrix")
print_matrix(A2, "Matrix A")
print_matrix(B2, "Matrix B")
if result2:
    print_matrix(result2, "Result (A @ B)")

# Test case 3: Invalid multiplication
A3 = [[1, 2, 3], [4, 5, 6]]
B3 = [[1, 2], [3, 4]]
print("Test 3: Invalid multiplication (2x3 @ 2x2)")
result3 = matrix_multiply_loops(A3, B3)
```

(i) Given:

```python
X = [1,2,3,4,5]
y = [2,4,5,4,5]
```

Compute:

mean of X

mean of y

slope (m)

intercept (c)

predicted values

No libraries allowed.(sklearn or numpy)


```python
# Q4: Linear Regression - Calculate slope and intercept without numpy/sklearn

X = [1, 2, 3, 4, 5]
y = [2, 4, 5, 4, 5]

# 1. Calculate mean of X
mean_X = sum(X) / len(X)
print(f"Mean of X: {mean_X}")

# 2. Calculate mean of y
mean_y = sum(y) / len(y)
print(f"Mean of y: {mean_y}")

# 3. Calculate slope (m) using formula: m = sum((xi - mean_x)(yi - mean_y)) / sum((xi - mean_x)^2)
numerator = 0
denominator = 0

for i in range(len(X)):
    numerator += (X[i] - mean_X) * (y[i] - mean_y)
    denominator += (X[i] - mean_X) ** 2

slope = numerator / denominator
print(f"Slope (m): {slope}")

# 4. Calculate intercept (c) using formula: c = mean_y - m * mean_x
intercept = mean_y - slope * mean_X
print(f"Intercept (c): {intercept}")

# 5. Calculate predicted values: y_pred = mx + c
predicted_y = []
for x_val in X:
    y_pred = slope * x_val + intercept
    predicted_y.append(y_pred)

print(f"\nPredicted values: {predicted_y}")
print(f"Actual values: {y}")

# Print regression line equation
print(f"\nLinear Regression Line: y = {slope:.4f}x + {intercept:.4f}")
```
