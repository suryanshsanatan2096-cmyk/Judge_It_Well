## Exercise 1

**Follow along with these steps**
* Create a figure object called fig using plt.figure()
* Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax.
* Plot (x,y) on that axes and set the labels and titles to match the plot below:

```python
# Exercise 1: Create figure with single axis and plot
# Import data first
import numpy as np
x = np.arange(0, 100)
y = x * 2
z = x ** 2

# Create figure object
fig = plt.figure()

# Add axis at [0,0,1,1] (full canvas)
ax = fig.add_axes([0, 0, 1, 1])

# Plot (x,y) on the axis
ax.plot(x, y)

# Set labels and title
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_title('title')

# Display the figure
fig
```

## Exercise 2
**Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively.**

```python
# Exercise 2: Create two axes at different positions
fig = plt.figure()

# Create ax1 at [0,0,1,1]
ax1 = fig.add_axes([0, 0, 1, 1])

# Create ax2 at [0.2, 0.5, .2, .2]
ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])

print("Two axes created successfully")
```

**Now plot (x,y) on both axes. And call your figure object to show it.**

```python
# Exercise 2 Part 2: Plot (x,y) on both axes and show the figure
# Plot on ax1 (main axis)
ax1.plot(x, y)

# Plot on ax2 (inset axis)
ax2.plot(x, y)

# Show the figure
fig
```

## Exercise 3

**Create the plot below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]**

```python
# Exercise 3: Create plot with two axes and zoom effect
fig = plt.figure()

# Add main axis at [0,0,1,1]
ax1 = fig.add_axes([0, 0, 1, 1])

# Add zoomed inset axis at [0.2, 0.5, 0.4, 0.4]
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])

# Plot z (quadratic) on main axis
ax1.plot(x, z)

# Plot on zoomed axis with limited range
ax2.plot(x, z)
ax2.set_xlim(0, 50)
ax2.set_ylim(0, 2500)

ax2.set_title('zoom')

fig
```

**Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
# Exercise 3 Part 2: Use x, y, z arrays with zoom effect
fig = plt.figure()

# Add main axis
ax1 = fig.add_axes([0, 0, 1, 1])

# Add zoomed inset
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])

# Plot z (quadratic) on main axis
ax1.plot(x, z, color='blue')
ax1.set_xlabel('X')
ax1.set_ylabel('Z')

# Plot on zoomed axis with focused range
ax2.plot(x, z, color='blue')
ax2.set_xlim(20, 22)
ax2.set_ylim(30, 50)
ax2.set_title('zoom')

fig
```

## Exercise 4

**Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
# Exercise 4 Part 1: Create subplots with 1 row and 2 columns
fig, axes = plt.subplots(nrows=1, ncols=2)
fig
```

**Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
# Exercise 4 Part 2: Plot (x,y) and (x,z) on the axes with different styles
fig, axes = plt.subplots(nrows=1, ncols=2)

# Plot (x,y) on first axis with dashed blue line
axes[0].plot(x, y, linestyle='--', linewidth=2, color='blue')
axes[0].set_title('(x,y)')

# Plot (x,z) on second axis with solid red line
axes[1].plot(x, z, linestyle='-', linewidth=2, color='red')
axes[1].set_title('(x,z)')

fig
```

**See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.**

```python
# Exercise 4 Part 3: Resize the plot using figsize argument
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 4))

# Plot (x,y) on first axis with dashed blue line
axes[0].plot(x, y, linestyle='-', linewidth=2, color='blue')
axes[0].set_title('(x,y) - Resized')

# Plot (x,z) on second axis with dashed red line
axes[1].plot(x, z, linestyle='--', linewidth=2, color='red')
axes[1].set_title('(x,z) - Resized')

fig
```

# #**Plots**

Plot the following plots using matplotlib, you many generate your own data or get it from the internet:

1. Histogram
2. Scatter Plot
3. Boxplot
4. Pie charts

(Bonus - Read up on the use-case of histograms in EDA and ML in general, they are VERY useful! Also, look up **Violin plots** if you have extra time)


```python
# Plot 1: Histogram
plt.figure(figsize=(14, 10))

# Generate random data for histogram
data = np.random.normal(loc=100, scale=15, size=1000)

plt.subplot(2, 2, 1)
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram - Distribution of Data')
plt.xlabel('Values')
plt.ylabel('Frequency')

# Plot 2: Scatter Plot
plt.subplot(2, 2, 2)
scatter_x = np.random.randn(100)
scatter_y = 2 * scatter_x + np.random.randn(100)
plt.scatter(scatter_x, scatter_y, alpha=0.6, color='coral', s=50)
plt.title('Scatter Plot')
plt.xlabel('X Values')
plt.ylabel('Y Values')

# Plot 3: Boxplot
plt.subplot(2, 2, 3)
box_data = [np.random.normal(0, std, 100) for std in range(1, 4)]
plt.boxplot(box_data, labels=['Group 1', 'Group 2', 'Group 3'])
plt.title('Boxplot - Distribution by Group')
plt.ylabel('Values')

# Plot 4: Pie Chart
plt.subplot(2, 2, 4)
sizes = [30, 25, 20, 25]
labels = ['Category A', 'Category B', 'Category C', 'Category D']
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']
explode = (0.05, 0, 0, 0)
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', explode=explode, startangle=90)
plt.title('Pie Chart - Proportion Distribution')

plt.tight_layout()
plt.show()
```
