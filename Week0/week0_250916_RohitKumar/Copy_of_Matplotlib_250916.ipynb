**Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([.2,.5,.4,.4])

x_main = np.arange(0,100)
y_main = x_main*2
z_main = x_main**2

# Plot on ax1
ax1.plot(x_main, z_main, color='b')
ax1.set_xlabel("X")
ax1.set_ylabel("Z")

ax1.set_xlim([0, 100])
ax1.set_ylim([0, 10000])


x_zoom = np.arange(20.0, 22.0,0.01)
y_zoom = x_zoom * 2

# Plot on ax2
ax2.plot(x_zoom, y_zoom, color='b')
ax2.set_xlabel("X")
ax2.set_ylabel("Y")


ax2.set_xticks(np.arange(20.0, 22.1, 0.5))
ax2.set_xlim([20, 22])
ax2.set_ylim([30, 50])
ax2.set_title("zoom")
```

## Exercise 4

**Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

ax1 = plt.subplot(1,2,1)
ax2 = plt.subplot(1,2,2)
```

**Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

ax1 = plt.subplot(1,2,1)
ax2 = plt.subplot(1,2,2)

x_zoom = np.arange(0,100)
y_zoom = x_zoom * 2

# Plot on ax1
ax1.plot(x_zoom, y_zoom, color='b',ls="--")
ax1.set_xticks(np.linspace(0,100,5))

ax1.set_xlim([-5, 105])
ax1.set_ylim([-5, 205])

x_main = np.arange(0,100)
z_main = x_main**2

# Plot on ax2
ax2.plot(x_main, z_main, color='r')



```

**See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.**

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(8, 2))
ax1 = plt.subplot(1,2,1)
ax2 = plt.subplot(1,2,2)

x_zoom = np.arange(0,100)
y_zoom = x_zoom * 2

# Plot on ax1
ax1.plot(x_zoom, y_zoom, color='b',ls="-",lw =3)
ax1.set_xticks(np.linspace(0,100,5))

ax1.set_xlim([-5, 105])
ax1.set_ylim([-5, 205])

x_main = np.arange(0,100)
z_main = x_main**2

# Plot on ax2
ax2.plot(x_main, z_main, color='r',ls="--",lw=3)



```
