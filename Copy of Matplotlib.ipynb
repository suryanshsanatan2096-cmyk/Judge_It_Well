# #**Plots**

Plot the following plots using matplotlib, you many generate your own data or get it from the internet:

1. Histogram
2. Scatter Plot
3. Boxplot
4. Pie charts

(Bonus - Read up on the use-case of histograms in EDA and ML in general, they are VERY useful! Also, look up **Violin plots** if you have extra time)


```python
import matplotlib.pyplot as plt

# Sample data for the pie chart
labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [15, 30, 45, 10] # Percentages or counts
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Category B')

# Create a figure and axes
fig, ax = plt.subplots()

# Plot the pie chart
ax.pie(sizes, explode=explode, labels=labels, colors=colors,
       autopct='%1.1f%%', shadow=True, startangle=90)
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

ax.set_title('Sample Pie Chart')

plt.show()
```

```python
#2
import matplotlib.pyplot as plt

# Create a figure and axes
fig, ax = plt.subplots()

# Plot the scatter plot
ax.scatter(x, y)

# Add labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Scatter Plot of X vs Y')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some sample data for boxplots
data1 = np.random.normal(0, 1, 100)
data2 = np.random.normal(1, 2, 100)
data3 = np.random.normal(-1, 0.5, 100)

# Combine data into a list
boxplot_data = [data1, data2, data3]

# Create a figure and axes
fig, ax = plt.subplots()

# Plot the boxplot
ax.boxplot(boxplot_data)

# Add labels and title
ax.set_xticklabels(['Data1', 'Data2', 'Data3'])
ax.set_ylabel('Value')
ax.set_title('Boxplot of Sample Data')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Create a figure with two subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))

# Histogram for x
axes[0].hist(x, bins=10, edgecolor='black', alpha=0.7)
axes[0].set_xlabel('X Value')
axes[0].set_ylabel('Frequency')
axes[0].set_title('Histogram of X')

# Histogram for y
axes[1].hist(y, bins=10, edgecolor='black', alpha=0.7)
axes[1].set_xlabel('Y Value')
axes[1].set_ylabel('Frequency')
axes[1].set_title('Histogram of Y')

plt.tight_layout() # Adjust layout to prevent overlapping titles/labels
plt.show()
```
